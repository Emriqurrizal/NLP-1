```python
import time

all_reviews = []
continuation_token = None
max_iterations = 100  # Batas aman untuk mencegah loop tak terbatas
iteration = 0

print("Starting to scrape reviews...")

while iteration < max_iterations:
    try:
        result, continuation_token = reviews(
            app_id,
            continuation_token=continuation_token,
            lang='en',
            country='us',
            sort=Sort.NEWEST,
            count=200
        )
        
        if result:
            all_reviews.extend(result)
            iteration += 1
            print(f"Iteration {iteration}: Scraped {len(result)} reviews | Total: {len(all_reviews)} reviews")
        else:
            print("No more reviews returned.")
            break
        
        # Berhenti jika tidak ada continuation token 
        if continuation_token is None:
            print("No continuation token - all reviews have been scraped!")
            break
        
        # Tambahkan jeda untuk menghindari rate limiting
        time.sleep(1)
        
    except Exception as e:
        print(f"Error during scraping: {str(e)}")
        print(f"Successfully scraped {len(all_reviews)} reviews before error")
        break

# Konversi ke DataFrame
df_all = pd.DataFrame(all_reviews)

# Pilih hanya kolom yang diperlukan 
columns_to_keep = ['reviewId', 'userName', 'content', 'score', 'thumbsUpCount', 
                   'reviewCreatedVersion', 'at', 'replyContent', 'repliedAt']

# Simpan hanya kolom yang ada di dataframe
columns_to_keep = [col for col in columns_to_keep if col in df_all.columns]
df_all = df_all[columns_to_keep]

# Simpan ke CSV
df_all.to_csv('../csv/stockbit_reviews.csv', index=False)
print(f"\n{'='*50}")
print(f"Scraping complete!")
print(f"Total reviews scraped: {len(df_all)}")
print(f"Total iterations: {iteration}")
print(f"Saved to: ../csv/stockbit_reviews.csv")
print(f"Columns: {list(df_all.columns)}")
print(f"{'='*50}")
```
